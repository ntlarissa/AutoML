```python
#fonction permenttant de diviser les données de tests de d'apprentissaged
def train_split(data):
    X_train=data.drop(['Label'], axis=1)
    y_train=data['Label']
    X_train, X_test, Y_train, Y_test = train_test_split(X_train, y_train, test_size=0.1)
    return X_test,Y_test

#lecture des données en supprimant les colonnes qui ne font pas partir de l'apprentissage
data = pd.read_csv("./Friday-02-03-2018_TrafficForML_CICFlowMeter.csv", skipinitialspace=True, quotechar='"').drop([ 'Dst Port', 'Protocol', 'Timestamp'], axis=1)
data['Flow Byts/s']=data['Flow Byts/s'].astype('float')
data['Flow Pkts/s']=data['Flow Byts/s'].astype('float')
data=data.replace([np.Inf, np.NINF], np.nan).dropna()

#parcours tous les fichiers
for x in glob('./CSV_CICIDS/*.csv'):
    print(x)
    
    data_train = pd.read_csv(x, skipinitialspace=True, quotechar='"').drop([ 'Dst Port', 'Protocol', 'Timestamp'], axis=1)
    data_train['Flow Byts/s']=data_train['Flow Pkts/s'].astype('float')
    data_train['Flow Pkts/s']=data_train['Flow Pkts/s'].astype('float')
    data_train=data_train.replace([np.Inf, np.NINF], np.nan).dropna()
    print(Counter(data_train['Label']))
    print(data_train.shape)
    data =pd.concat([data,data_train],axis=0)
    
```
