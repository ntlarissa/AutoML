```python
def train_split(data,txval,txtest):
    X_train=data.drop(['attack_cat'], axis=1)
    y_train=data['attack_cat']
    X_train1, X_test, Y_train1, Y_test = train_test_split(X_train, y_train, test_size=txtest)
    X_train, X_val, Y_train, Y_val = train_test_split(X_train1, Y_train1, test_size=txval)
    return X_train,Y_train, X_val, Y_val, Y_train, Y_test

data_train = pd.read_csv('./UNSW_NB15_training-set_selected.csv').drop([ 'id', 'label'], axis=1)

X_train,Y_train, X_val, Y_val, X_test, Y_test = train_split(data_train,0.2,0.1)

#print(Counter(Y))

#X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1)

print(Counter(Y_train))
print(Counter(Y_val))
print(Counter(Y_test))
```
